**Describe the linear search solution explained above problem in your own words.**
1. Create a variable position with value 1
2. Compare the number at current position to the number before it
3. If the number is smaller than its predecessor, then return position.
4. Otherwise, increment position and repeat till we exhaus all the numbers.

```python
def count_rotations_linear(nums):
    position = 0 # for now use 0 to easier understand the code

    while position < len(nums):
        if position > 0 and nums[position] < nums[position-1]:
            return position

        position += 1

    return 0
```

## Binary Search Solution

**Describe the binary search solution explained above problem in your own words.**
1. Find the mid element
2. Check if the mid element is smaller than left then return its position
3. Check if the mid element is smaller than the last element of the range then go find first half of the list
4. Check if the mid element is bigger than the last element of the range then go find second half of the list
5. If no more elements return 0

```python
def count_rotations_binary(nums):
    lo = 0
    hi = len(nums) - 1

    while lo <= hi:
        mid = (lo + hi) // 2
        mid_number = nums[mid]

        if mid > 0 and mid_number < nums[mid-1]:
            return mid
        elif mid_number < nums[hi]:
            hi = mid - 1
        else:
            lo = mid + 1
    
    return 0
```
